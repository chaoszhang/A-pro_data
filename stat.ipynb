```python
%matplotlib inline
import math
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 8]

cat = []
val = []
vcat = [[] for i in range(50)]
with open("../e4_0_70_ils.stat", "r") as ins:
    n = 0
    for line in ins:
        arr = line.split()
        if len(arr) > 2 and int(arr[0]) >= 12:
            cat.append(math.floor(n/1000))
            val.append(float(arr[2]) * 100)
            vcat[int(math.floor(n/1000))].append(float(arr[2]) * 100)
        n = n + 1
#print [np.mean(l) for l in vcat]
#print max([np.mean(l) for l in vcat]), min([np.mean(l) for l in vcat])
#print max(val), min(val)
plot = sns.violinplot(x=cat, y=val, order=np.argsort([np.mean(l) for l in vcat]), cut=0)
#plot.axes.get_xaxis().set_visible(False)
plot.axes.get_xaxis().set_ticks([])
plt.xlabel("Replicates")
plt.ylabel("Distribution of genetree ILS level (%)")
```

```python
cat = []
val = []
vcat = [[] for i in range(50)]
h = []
with open("../default_duploss.stat", "r") as ins:
    for line in ins:
        arr = line.split()
        cat.append(int(arr[0])-1)
        val.append(int(arr[1]))
        h.append("Duplication")
        cat.append(int(arr[0])-1)
        val.append(int(arr[2]))
        h.append("Loss")
        vcat[int(arr[0])-1].append(int(arr[1]))
plot = sns.violinplot(x=cat, y=val, hue=h, order=np.argsort([np.mean(l) for l in vcat]), cut=0, split=True, inner="quartile")
#plot.axes.get_xaxis().set_visible(False)
plot.axes.get_xaxis().set_ticks([])
plt.xlabel("Replicates")
plt.ylabel("Distribution of # of duplication and loss events in genetrees")
```
